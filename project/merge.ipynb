```python
import pandas as pd
```

```python
accidents = pd.read_json("accidents.json")
```

```python
beats = pd.read_csv("ChicagoAreasBeatDistricts.csv")
beats
```

```python
beats_clean = beats.drop_duplicates(subset='Beat', keep="first")
beats_clean
```

```python
accidents.rename(columns={"beat_of_occurrence": "Beat"}, inplace = True)
```

```python
accidents.describe
```

```python
accidents_clean = accidents.drop(columns=['crash_record_id', 'location', 'injuries_unknown', 'injuries_incapacitating', 'injuries_fatal', 'injuries_no_indication', 'injuries_reported_not_evident', 'injuries_non_incapacitating', 'private_property_i', 'crash_date_est_i', 'statements_taken_i', 'photos_taken_i', 'work_zone_i', 'work_zone_type', 'workers_present_i', 'dooring_i'])
```

```python
accidents_clean.drop(columns=['crash_hour', 'crash_day_of_week', 'crash_month', 'traffic_control_device', 'posted_speed_limit', 'device_condition'], inplace=True)
```

```python
accidents_clean.drop(columns=['first_crash_type', 'trafficway_type', 'weather_condition', 'num_units'], inplace=True)
```

```python
accidents_clean.drop(columns=['alignment', 'lighting_condition', 'roadway_surface_cond', 'road_defect', 'report_type', 'crash_type', 'intersection_related_i', 'damage'], inplace=True)
```

```python
accidents_clean.rename(columns={"beat_of_occurrence": "Beat"}, inplace = True)
```

```python
accidents_clean.drop(columns=['date_police_notified', 'prim_contributory_cause', 'sec_contributory_cause', 'street_direction'], inplace=True)
```

```python
accidents_clean.drop(columns=['most_severe_injury', 'injuries_total', 'rd_no', 'lane_cnt'], inplace=True)
```

```python
accidents_clean
```

```python
merged_clean = pd.merge(accidents_clean, beats_clean, on='Beat', how='inner')
merged_clean
```
